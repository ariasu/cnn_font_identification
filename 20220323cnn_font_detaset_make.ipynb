```python
DATADIR = r"D:\application\OneDrive\デスクトップ\AnaChro\cnn_font_jpeg"
CATEGORIES = ["gothic", "mincho","kyokasyo"]
IMG_SIZE = 64
training_data = []

for class_num, category in enumerate(CATEGORIES):
    path = os.path.join(DATADIR, category)
    os.chdir(path)  # 相対パス
    print(os.getcwd())
    gif = os.listdir(path)
    
    for i in range(len(os.listdir(path))):        
        img_array = imread(gif[i])
        img_array_resize = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        training_data.append([img_array_resize, class_num])
    
random.shuffle(training_data)

X_train = []  # 画像データ
y_train = []  # ラベル情報

# データセット作成
for feature, label in training_data:
    X_train.append(feature)
    y_train.append(label)
    
# numpy配列に変換
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
# データセットの確認
for i in range(0, 4):
    print("学習データのラベル：", y_train[i])
    plt.subplot(2, 2, i+1)
    plt.axis('off')
    plt.title(label = 'gothic' 
              if y_train[i] == 0 
              elif y_train[i] == 1
              label = 'mincho'
              else "kyokasyo")
    plt.imshow(X_train[i], cmap='gray')
plt.show()
```

```python
np.save(os.path.join(DATADIR, 'cnn_font_gif'), X_train)
np.save(os.path.join(DATADIR, 'cnn_font_label'), y_train)
```
