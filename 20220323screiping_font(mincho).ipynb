```python
import numpy as np
from selenium import webdriver as web
import io
from urllib import request
import os
from PIL import Image
import time
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = web.Chrome('./chromedriver')
url = 'https://www.syotai.net/mincho/'
driver.get(url)
```

```python
os.chdir('.\cnn_font_jpeg\mincho')  # 相対パス
print(os.getcwd())
#/html/body/div[1]/div[2]/div/main/article/div/p[2]/a[1]/img
```

```python
imgs = []
i = 1

for v in range(2,45):
    i = 1
    try:
        while(i < 100):
            a = '/html/body/div[1]/div[2]/div/main/article/div/p[{0}]/a[{1}]/img'.format(v,i)
            yes = driver.find_element_by_xpath(a)
            ActionChains(driver).move_to_element(yes).perform()# 要素を表示するようスクロール
            img = yes.get_attribute('src')
            imgs.append(img)
            i += 1
    except Exception:
        print('')
```

```python
print(len(imgs))
print(imgs[1238])
```

```python
start = 1238
stop = len(imgs)
for c in range(start,stop):
    print(c)
```

```python
start = 1238
stop = len(imgs)
for c in range(start, stop):
    f = io.BytesIO(request.urlopen(imgs[c]).read())
    img_open = Image.open(f)
    img_open.save('mincho' + str(c-1) + '.gif')
```
